```python
import pandas as pd
import json
import os

def read_emotion_json_files(emolist, base_path='../data/label/train'):
    # 감정별 JSON 파일을 읽고 DataFrame으로 변환하는 딕셔너리
    emotion_dataframes = {}
    
    for emotion in emolist:
        file_path = os.path.join(base_path, f'train_{emotion}.json')
        
        # euc-kr 인코딩으로 JSON 파일 열기 및 읽기
        try:
            with open(file_path, encoding='euc-kr') as file:
                data = json.load(file)
                
            # JSON 데이터를 DataFrame으로 생성
            df = pd.json_normalize(data)
            emotion_dataframes[emotion] = df
        
        except FileNotFoundError:
            print(f'파일을 찾을 수 없습니다: {file_path}')
        except json.JSONDecodeError:
            print(f'JSON 파일을 디코드하는 중 오류가 발생했습니다: {file_path}')
    
    return emotion_dataframes

# 감정 리스트
emolist = ['anger', 'happy', 'panic', 'sadness']

# 감정별 JSON 파일을 읽고 DataFrame으로 변환
emotion_dataframes = read_emotion_json_files(emolist)

# 각 감정에 해당하는 DataFrame 출력
for emotion, df in emotion_dataframes.items():
    print(f"{emotion} 데이터프레임:")
    globals()[f"{emotion}_df"] = df
    display(df)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def compare_face_expressions(df):
    match_counts = 0
    mismatch_counts = {
        '1_mismatch': 0,
        '2_mismatch': 0,
        '3_mismatch': 0
    }
    
    for _, row in df.iterrows():
        faceExp_uploader = row['faceExp_uploader']
        annot_A_faceExp = row['annot_A.faceExp']
        annot_B_faceExp = row['annot_B.faceExp']
        annot_C_faceExp = row['annot_C.faceExp']
        
        mismatches = 0
        if faceExp_uploader != annot_A_faceExp:
            mismatches += 1
        if faceExp_uploader != annot_B_faceExp:
            mismatches += 1
        if faceExp_uploader != annot_C_faceExp:
            mismatches += 1
        
        if mismatches == 0:
            match_counts += 1
        elif mismatches == 1:
            mismatch_counts['1_mismatch'] += 1
        elif mismatches == 2:
            mismatch_counts['2_mismatch'] += 1
        elif mismatches == 3:
            mismatch_counts['3_mismatch'] += 1
    
    return match_counts, mismatch_counts

# 각 감정 데이터프레임에 대해 함수 호출 및 결과 저장
results = {
    'emotion': [],
    'match': [],
    '1_mismatch': [],
    '2_mismatch': [],
    '3_mismatch': []
}

for emo, name in zip(emo_df_list, emo_names):
    match_counts, mismatch_counts = compare_face_expressions(emo)
    results['emotion'].append(name)
    results['match'].append(match_counts)
    results['1_mismatch'].append(mismatch_counts['1_mismatch'])
    results['2_mismatch'].append(mismatch_counts['2_mismatch'])
    results['3_mismatch'].append(mismatch_counts['3_mismatch'])

# 결과를 DataFrame으로 변환
results_df = pd.DataFrame(results)

# 막대그래프 그리기
bar_width = 0.2
index = np.arange(len(results_df))

fig, ax = plt.subplots(figsize=(10, 6))

bars1 = ax.bar(index, results_df['match'], bar_width, label='Match')
bars2 = ax.bar(index + bar_width, results_df['1_mismatch'], bar_width, label='1 Mismatch')
bars3 = ax.bar(index + 2 * bar_width, results_df['2_mismatch'], bar_width, label='2 Mismatches')
bars4 = ax.bar(index + 3 * bar_width, results_df['3_mismatch'], bar_width, label='3 Mismatches')

# 수치 표시
for bars in [bars1, bars2, bars3, bars4]:
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval + 0.05, int(yval), ha='center', va='bottom')

plt.xlabel('Emotion')
plt.ylabel('Counts')
plt.title('Face Expression Matching Results by Emotion')
plt.xticks(index + 1.5 * bar_width, results_df['emotion'])
plt.legend()

plt.show()
```
